```python
"""
Standalone WAV File Player
Loads and plays any WAV file from the wav/ folder
"""

from pathlib import Path
from ipywidgets import Dropdown, Output, VBox, Button, HBox, HTML
from IPython.display import Audio, display, clear_output

# ============================================================================
# WAV FILE BROWSER & PLAYER
# ============================================================================

def create_wav_player():
    """Standalone player for browsing and playing WAV files"""
    
    WAV_DIR = Path("wav")
    
    if not WAV_DIR.exists():
        print("âš ï¸ wav/ folder not found. Generate some audio files first!")
        return
    
    # Get all WAV files
    wav_files = sorted(WAV_DIR.glob("*.wav"))
    
    if not wav_files:
        print("âš ï¸ No WAV files found in wav/ folder. Generate some audio first!")
        return
    
    # Create file dropdown
    file_options = [(f.name, f) for f in wav_files]
    file_dropdown = Dropdown(
        options=file_options,
        description="Audio File:",
        layout={"width": "500px"},
    )
    
    # Info display
    info_label = HTML(value="")
    
    # Output area for player
    output = Output()
    
    # Refresh button
    refresh_button = Button(
        description="ðŸ”„ Refresh",
        tooltip="Reload file list",
        layout={"width": "100px"},
    )
    
    def get_file_info(filepath: Path) -> str:
        """Get file size and info"""
        size_kb = filepath.stat().st_size / 1024
        return f"ðŸ“Š Size: {size_kb:.1f} KB"
    
    def play_file(filepath: Path):
        """Play selected WAV file"""
        with output:
            clear_output(wait=True)
            
            # Update info
            info_label.value = get_file_info(filepath)
            
            # Display audio player
            print(f"ðŸŽµ Now playing: {filepath.name}")
            display(Audio(filename=str(filepath), autoplay=True))
    
    def on_file_change(change):
        if change["name"] == "value":
            play_file(change["new"])
    
    def on_refresh_clicked(_):
        """Refresh the file list"""
        nonlocal wav_files, file_options
        
        wav_files = sorted(WAV_DIR.glob("*.wav"))
        file_options = [(f.name, f) for f in wav_files]
        
        # Update dropdown options
        current_file = file_dropdown.value if file_dropdown.value in wav_files else None
        file_dropdown.options = file_options
        
        if current_file and current_file in wav_files:
            file_dropdown.value = current_file
        
        print(f"âœ“ Refreshed: {len(wav_files)} files found")
    
    # Attach event handlers
    file_dropdown.observe(on_file_change)
    refresh_button.on_click(on_refresh_clicked)
    
    # Initial playback
    if file_options:
        play_file(file_options[0][1])
    
    # Layout
    controls = HBox([file_dropdown, refresh_button])
    display(VBox([
        HTML(value="<h3>ðŸŽ§ WAV File Player</h3>"),
        controls,
        info_label,
        output
    ]))

# ============================================================================
# USAGE
# ============================================================================

print("ðŸ“» WAV File Player ready!")
print("Run: create_wav_player()\n")
```

```python
create_wav_player()
```
